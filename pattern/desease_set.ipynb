```python
import pandas as pd
```

```python
patient = pd.read_pickle('patient.pkl.zip')
patient.head()
```

```python
record = pd.read_pickle('record.pkl.zip')
record.head()
```

```python
data = pd.merge(patient, record, left_on='ID', right_on='PatientID')
data.head()
```

```python
groupby = data.set_index('ID')['ICD9CM'].groupby(level=0)
```

```python
desease_set = groupby.apply(lambda s: set(s.explode()))
desease_set.head()
```

```python
desease_set.to_pickle('disease_sets.pkl.zip')
```
